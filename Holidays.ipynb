#Grafico numero de feriados por mes

````python
import matplotlib.pyplot as plt

# Contar feriados por mes
feriados_mes = holidays["Month"].value_counts().sort_index()

# Gráfico de barras
plt.figure()
plt.bar(feriados_mes.index, feriados_mes.values)

plt.title("Número de feriados por mes")
plt.xlabel("Mes")
plt.ylabel("Cantidad de feriados")

plt.xticks(range(1,13))
plt.show()
````

Segun los graficos los meses con mayor numero de feriados no tienen mayor numero de pasajefrons, podemos ver que en la grafica de numero de feriados por mes el que mas cantidad de feriads tiene es el mes 12, y al compararla con la grafica de mes vs promedio de pasajeros el mes 12 no tienen mayor cantidad de pasajeros.


##PUNTOS EXTRA
a) Los países con mayor número total de festivos.

````python
import matplotlib.pyplot as plt

# Top 10 países con más festivos
top10 = holidays["ADM_name"].value_counts().sort_values(ascending=False).head(10)

plt.figure()
plt.bar(top10.index, top10.values)
plt.xticks(rotation=45)
plt.title("Top 10 países con mayor número de festivos")
plt.xlabel("País")
plt.ylabel("Número de festivos")
plt.show()
````

Elegi el grafico de barras porque me permite observar mejor y tener los datos mas organizados de que paises tienen mas feriados y  con que cantidad total de feriados. Por otro lado decidi poner el top 10 para que se viera mas organizado, ya que cuando los ponia todo no eran legibles los paises

b) La concentración mensual global.


````python
# Total de festivos por mes
festivos_mes = holidays["Month"].value_counts().sort_index()

plt.figure()
plt.bar(festivos_mes.index, festivos_mes.values)
plt.title("Concentración global de festivos por mes")
plt.xlabel("Mes")
plt.ylabel("Número de festivos")
plt.xticks(range(1,13))
plt.show()
````

Elegi grafico de barras porque puedo comparar facilmente en que meses hay mas festivos a nivel global, es una manera sencilla para analizar.
